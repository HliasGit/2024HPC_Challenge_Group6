```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "../io/execution_time.csv"
data = pd.read_csv(path, names=['NumThreads', 'ExecutionTime'])
data
```

```python
execution_time_df = data.groupby('NumThreads')['ExecutionTime'].mean().reset_index()
execution_time_df
```

```python
sns.set_theme(style="darkgrid")

x_label = execution_time_df.columns[0]
y_label = execution_time_df.columns[1]

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.lineplot(x=x_label, y=y_label, data=execution_time_df, marker='o', ax=axes[0])
axes[0].set_title('Execution time')
axes[0].set_xlabel(x_label)
axes[0].set_ylabel(y_label + ' [s]')

sns.lineplot(x=x_label, y=y_label, data=execution_time_df.iloc[:6], marker='o', ax=axes[1])
axes[1].set_title('Execution time')
axes[1].set_xlabel(x_label)
axes[1].set_ylabel(y_label + ' [s]')

# Regola automaticamente il layout per evitare sovrapposizioni
plt.tight_layout()

# Mostra la figura
plt.show()
```
