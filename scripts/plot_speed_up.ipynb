```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "../io/execution_time.csv"
data = pd.read_csv(path, names=['NumThreads', 'ExecutionTime'])
data
```

```python
execution_time_df = data.groupby('NumThreads')['ExecutionTime'].mean().reset_index()
execution_time_df
```

```python
x_label = execution_time_df.columns[0]
y_label = execution_time_df.columns[1]

plt.figure(figsize=(8, 6))
sns.lineplot(x=x_label, y=y_label, data=execution_time_df, marker='o')
plt.title('Execution time')
plt.xlabel(x_label)
plt.ylabel(y_label + ' [ms]')
plt.show()

plt.figure(figsize=(8, 6))
sns.lineplot(x=x_label, y=y_label, data=execution_time_df[0:5], marker='o')
plt.title('Execution time')
plt.xlabel(x_label)
plt.ylabel(y_label + ' [ms]')
plt.show()
```
